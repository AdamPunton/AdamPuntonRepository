```python
df = df.sort_index(ascending=True)
df
```

```python
column_dfs = {}

for i in range(0, len(df.columns)):
    column_name = df.columns[i]
    
    column_df = pd.DataFrame({column_name: df.iloc[:, i]})
    
    column_dfs[column_name] = column_df

for column, column_df in column_dfs.items():
    print(column_df)
    print()
```

```python
stock_cumulative_returns = pd.concat([(1 + df[column]).cumprod() * 1000 for column, df in column_dfs.items()], axis=1)


folder = "Trend Portfolio Graphics"
try:
    os.mkdir(folder)
except FileExistsError:
    print(folder, "already exists")

pp = PdfPages(folder + "/compounding_return.pdf")
fig = plt.figure()
fig.suptitle('Compounding Return')
plt.plot(stock_cumulative_returns)
plt.legend(stock_cumulative_returns.columns)
plt.xlabel('Year')
plt.ylabel('Cumulative Return (Initial Investment = $1000)')
pp.savefig(fig)
pp.close()
```

```python
stocks = ['AAPL', 'AMZN', 'MSFT', 'BRK-B', 'GPK', 'XOM', '^GSPC']
market = '^GSPC'

start_date = '1999-01-01'
end_date = '2023-12-31'

data = yf.download(stocks + [market], start=start_date, end=end_date)['Adj Close']

returns_table = data.pct_change().resample('Y').sum()

stock_returns_table = returns_table[stocks]

market_returns = returns_table[market]

correlation_matrix = stock_returns_table.corrwith(market_returns)

print(correlation_matrix)
```

# Trend Portfolio Return Projection

```python
stocks = ['AAPL', 'AMZN', 'MSFT', 'BRK-B', 'GPK', 'XOM']
weights = [0.1, 0.1, 0.1, 0.35, 0.1, 0.15]

start_date = '1998-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')
stock_data = pd.DataFrame()

for ticker in stocks:
    data = pd.read_csv(f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={int(datetime.strptime(start_date, "%Y-%m-%d").timestamp())}&period2={int(datetime.strptime(end_date, "%Y-%m-%d").timestamp())}&interval=1mo&events=history&includeAdjustedClose=true')
    data.set_index('Date', inplace=True)
    data.index = pd.to_datetime(data.index)
    annual_returns = data['Adj Close'].resample('Y').last().pct_change().dropna()
    stock_data[ticker] = annual_returns

stock_returns = stock_data.dropna()
stock_returns
```

```python
trend_port_return_data = pd.read_excel('Trend_portfolio_return_data.xlsx')
trend_port_return_data
```

```python
trend_port_return_data.columns
```

```python
weights = [0.1, 0.1, 0.1, 0.35, 0.1, 0.15]

cc_return_metrics = trend_port_return_data[['Chesepeake Capital',
                                         'AAPL', 'AMZN', 'MSFT',
                                         'BRK-B', 'GPK']]

cc_portfolio_returns = cc_return_metrics.dot(weights)

cc_portfolio_returns
```

```python
weights = [0.1, 0.1, 0.1, 0.35, 0.1, 0.15]

xom_return_metrics = trend_port_return_data[['XOM',
                                         'AAPL', 'AMZN', 'MSFT',
                                         'BRK-B', 'GPK']]

xom_portfolio_returns = xom_return_metrics.dot(weights)
xom_portfolio_returns
```

```python
folder = "Trend Portfolio Graphics"
try:
    os.mkdir(folder)
except FileExistsError:
    print(folder, "already exists")


years = np.arange(1999, 2023)  
plt.plot(years, cc_portfolio_returns, label='Chesepeake Capital Portfolio')
plt.plot(years, xom_portfolio_returns, label='XOM Portfolio')
plt.xlabel('Year')
plt.ylabel('Portfolio Returns')
plt.title('Portfolio Returns Comparison')
plt.legend()
plt.savefig(folder + "/XOM vs. Chesepeake.jpg")

plt.show()
```

```python
difference = cc_portfolio_returns - xom_portfolio_returns
percentage_difference = difference * 100

print("Chesepeake Capital vs. XOM")
for diff in percentage_difference:
    print(f"{diff:.2f}%")
```

```python
average_return_comparison = sum(difference) / len(difference)
average_return_comparison = average_return_comparison * 100

print("Chesepeake Capital vs. XOM Average Return")
print(f"{average_return_comparison:.2f}%")
```
