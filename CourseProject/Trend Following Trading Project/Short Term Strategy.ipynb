```python
import yfinance as yf
import pandas as pd
from datetime import datetime

tickerSymbol = 'CL=F'

tickerData = yf.Ticker(tickerSymbol)

tickerDf = tickerData.history(period='1d', start='2022-1-1', end='2023-4-21')

weekly_min_max = tickerDf.resample('M').agg({'Close': ['min', 'max']})
weekly_min_max
```

```python
today_close = tickerDf.iloc[-1].loc['Close']
print(today_close)
```

```python
min_prices = weekly_min_max['Close']['min'].tolist()
less_than_current = [price for price in min_prices if price < today_close]

max_prices = weekly_min_max['Close']['max'].tolist()
greater_than_current = [price for price in max_prices if price > today_close]

closest_min = min(less_than_current, key=lambda x: abs(x-today_close))
closest_max = min(greater_than_current, key=lambda x: abs(x-today_close))

print("Closest min value:", closest_min)
print("Current value:", today_close)
print("Closest max value:", closest_max)

```
